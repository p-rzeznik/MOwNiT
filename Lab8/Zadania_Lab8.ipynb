```julia
using Roots  
using Plots
using ForwardDiff
using Plots
```

```julia
#funkcja obliczajaca pochodna
D(f) = x->ForwardDiff.derivative(f, float(x))

```

```julia
function test(f, interval)
    
    p=plot(f, interval[1], interval[2])

    print(iszero(f(find_zero(f, (0, 1), Bisection(),verbose=true))) , "\n\n")
    print(iszero(f(find_zero((f, D(f)),0, Roots.Newton(),verbose=true))) , "\n\n")
    print(iszero(f(find_zero(f, 0,   verbose=true))) , "\n\n")
    
    return p
end
```

```julia
using DataFrames
df = DataFrame(func=[], method=[], iterations=[], evaluations=[], is_correct=[]);
```

## f(x) = sin( x ) - x / 2

```julia
f(x) = sin( x ) - x / 2
i = (-2,2)
plot(f, i[1], i[2])
```

```julia
test(f, i)
```

```julia
insert!.(eachcol(df), 1, ["sin( x ) - x / 2", "Bisection", 0, 2, true])
insert!.(eachcol(df), 1, ["sin( x ) - x / 2", "Newton", 0, 1, true])
insert!.(eachcol(df), 1, ["sin( x ) - x / 2", "Secant", 0, 2, true]);
```
